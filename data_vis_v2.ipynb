```python
from helpers.preprocessing import read_all_data
import matplotlib.pyplot as plt
```

```python
data_dict = read_all_data()
ann = data_dict['ann']
ann = ann['labels'].tolist()
```

```python
percent = lambda x: round((ann.count(x)/len(ann))*100, 2)
plt.title('Total class distribution {0}% 0, {1}% 1, {2}% 2, {3}% 3'.format(percent(0), percent(1), percent(2),
                                                                           percent(3)))
plt.grid()
plt.hist(ann)
plt.show()
```
