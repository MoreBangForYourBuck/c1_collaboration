```python
from helpers.preprocessing import read_all_data, get_distribution
import matplotlib.pyplot as plt
```

```python
data_dict = read_all_data()
ann = data_dict['ann']
ann = ann['labels'].tolist()
dist_dict = get_distribution(ann)
```

```python
percent = lambda x: round(x*100, 2)
plt.title('Total class distribution {0}% 0, {1}% 1, {2}% 2, {3}% 3'.format(*[percent(x) for x in dist_dict['fracs']]))
plt.grid()
plt.hist(ann)
plt.show()
```
